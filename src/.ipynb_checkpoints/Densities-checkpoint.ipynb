```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
from scipy.stats import gaussian_kde
```

```python
def map_kde(samples, method='normal'):
    kde = gaussian_kde(samples, bw_method=method)
    x_values = np.linspace(min(samples), max(samples), 1000)
    kde_values = kde.evaluate(x_values)
    mode_index = np.argmax(kde_values)
    mode_value = x_values[mode_index]
    return mode_value

def map_mcmc(samples):
    nums, ints = np.histogram(samples, bins=50)
    max_num = max(nums)
    idx = list(nums).index(max_num)
    mode = (ints[idx+1]+ints[idx])/2
    print(mode, '+/-', (ints[idx+1]-ints[idx])/2)
    return mode 
    
```
